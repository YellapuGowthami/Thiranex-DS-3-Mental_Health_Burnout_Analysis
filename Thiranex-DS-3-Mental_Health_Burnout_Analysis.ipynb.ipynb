```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("/content/mental_health_burnout_tech_2026.csv")
```

```python
# Display First 5 Rows

print("First 5 Rows:")
print(df.head())
```

```python
# Dataset Information

print("\nDataset Information:")
print(df.info())
```

```python
# Statistical Summary

print("\nStatistical Summary:")
print(df.describe())
```

```python
# Missing Values

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)
```

```python
# ----------------------------
# Visualization 1
# Burnout Level Distribution
# ----------------------------

plt.figure(figsize=(8,5))

sns.countplot(x='burnout_level', data=df)

plt.title("Burnout Level Distribution")

plt.show()
```

```python
# ----------------------------
# Visualization 2
# Stress Score Distribution
# ----------------------------

plt.figure(figsize=(8,5))

sns.histplot(df['stress_score'], bins=20)

plt.title("Stress Score Distribution")

plt.show()
```

```python
# ----------------------------
# Visualization 3
# Sleep vs Stress
# ----------------------------

plt.figure(figsize=(8,5))

sns.scatterplot(
    x='sleep_hours_per_night',
    y='stress_score',
    data=df
)

plt.title("Sleep Hours vs Stress Score")

plt.show()
```

```python
# ----------------------------
# Visualization 4
# Work Hours vs Burnout
# ----------------------------

plt.figure(figsize=(8,5))

sns.boxplot(
    x='burnout_level',
    y='work_hours_per_week',
    data=df
)

plt.title("Work Hours vs Burnout")

plt.show()
```

```python
# ----------------------------
# Visualization 5
# Correlation Heatmap
# ----------------------------

plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

# Key Insights

1. Employees with higher work hours tend to experience greater burnout.

2. Lower sleep duration is associated with higher stress scores.

3. Work-life balance significantly impacts employee mental health.

4. Burnout levels vary across different work modes.

5. Correlation analysis helped identify relationships between stress, burnout, and sleep patterns.
